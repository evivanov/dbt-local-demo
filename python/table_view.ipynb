```python
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine("postgresql+psycopg2://dbt:dbt@localhost:5432/dbt_demo")
```

```python
# Find available tables
tables = pd.read_sql(
    """
    SELECT table_schema, table_name
    FROM information_schema.tables
    WHERE table_schema NOT IN ('pg_catalog', 'information_schema');
    """,
    engine
)
print(tables)
```

```python
# Read a table into a dataframe
df = pd.read_sql("SELECT * FROM analytics.sales_summary LIMIT 10;", engine)
print(df)
```

```python
# Read a table into a dataframe
df2 = pd.read_sql("SELECT * FROM analytics.avg_prices", engine)
print(df2)
```
